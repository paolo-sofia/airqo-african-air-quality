```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import polars as pl
from dotenv import load_dotenv

load_dotenv()
```

```python
dataframe: pl.DataFrame = pl.read_csv(os.getenv("TRAIN_PATH"))
dataframe.head()
```

```python
dataframe = dataframe.with_columns(
    pl.col("date").str.to_datetime(format="%Y-%m-%d") + pl.duration(hours=pl.col("hour"))
)
dataframe = dataframe.drop([col for col in dataframe.columns if "uvaerosollayerheight" in col] + ["hour", "city", "country", "id", "month"])
dataframe.head()
```

```python
# for column in dataframe.select_dtypes(include="number").columns:
#     plot = dataframe[column].plot(kind="hist", bins=256, title=column)
#     plt.show()
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import make_column_transformer, ColumnTransformer


col_transformer: ColumnTransformer = make_column_transformer(
    (OrdinalEncoder(dtype=np.int32), ["site_id"]),
    remainder="passthrough",
    n_jobs=-1,
    verbose=1,
    verbose_feature_names_out=False
)

col_transformer.set_output(transform="polars")
dataframe = col_transformer.fit_transform(dataframe)
dataframe.head()
```

```python
from sklearn.impute import KNNImputer

knn_imputer: KNNImputer = KNNImputer(
    add_indicator=False
)

col_transformer: ColumnTransformer = make_column_transformer(
    (knn_imputer, dataframe.select(pl.col(pl.NUMERIC_DTYPES)).columns),
    remainder="passthrough",
    n_jobs=-1,
    verbose=1,
    verbose_feature_names_out=False
)

col_transformer.set_output(transform="polars")

dataframe: pl.DataFrame = col_transformer.fit_transform(dataframe)
```

```python
dataframe.head()
```

```python
dataframe.with_columns(
    pl.col("date").dt.ordinal_day().alias("cos_day_of_year")
)
```

```python
dataframe = dataframe.with_columns(
    pl.col("date").dt.month().alias("month"),
    # pl.col("date").dt.hour().alias("hour"),
    pl.col("date").dt.quarter().alias("quarter"),
    (pl.col("date").dt.ordinal_day() / 12).sin().alias("sin_day_of_year"),
    (pl.col("date").dt.ordinal_day() / 12).cos().alias("cos_day_of_year")
)
```

```python
dataframe = dataframe.with_row_index(name="index").drop("date")
```

```python
from sklearn.model_selection import KFold

gkf: KFold = KFold(
    shuffle=True,
    random_state=666
)

train_index, test_index = next(gkf.split(
    X=dataframe.drop(["site_id"]),
    y=dataframe.select("pm2_5"),
    groups=dataframe.select("site_id")
))
```

```python
X_train = dataframe.filter(pl.col("index").is_in(train_index)).drop(["index", "site_id"])
y_train = X_train.select("pm2_5")
X_train = X_train.drop(["pm2_5"])


X_test = dataframe.filter(pl.col("index").is_in(train_index)).drop(["index", "site_id"])
y_test = X_test.select("pm2_5")
X_test = X_test.drop(["pm2_5"])
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor: RandomForestRegressor = RandomForestRegressor(
    n_jobs=-1,
    warm_start=False,
    random_state=666,
    verbose=1
)

regressor.fit(X_train, y_train)
```

```python
regressor.score(X=X_test, y=y_test)
```

```python
from sklearn.metrics import root_mean_squared_error

root_mean_squared_error(y_true=y_test, y_pred=regressor.predict(X_test))
```

```python
importances: pd.DataFrame = pd.DataFrame(data={"column": regressor.feature_names_in_, "importance": regressor.feature_importances_}).sort_values("importance", ascending=False)
importances["cumsum"] = importances.importance.cumsum()
importances
```
